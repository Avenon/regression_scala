### 1. Загрузите данные из файла Dataset.csv

```scala
val dataFile = io.Source.fromFile("../data/raw/Dataset.csv")
```

```scala
val dataArray = dataFile.getLines.drop(1)
    .map(_.split(",").map(_.trim))
    .map { line => line.map { elem => 
                    elem match {
                      case "" => 0.0
                      case x => x.toDouble
                    }
                  }
                }.toArray
```

```scala
val dataMatrix = DenseMatrix(dataArray: _*)
```

```scala
dataMatrix.rows
dataMatrix.cols
```

__Посмотрим на первые 5 записей в этом наборе данных.__

```scala
dataMatrix(0 to 4, ::)
```

**Создадим массивы X из столбцов наших признаков и y - из столбца output значения которого будем предсказывать.**

```scala
val X = dataMatrix(::, 0 to 26)
```

```scala
val y = dataMatrix(::, 27)
```

__Отмасштабируем столбцы матрицы X, вычтя из каждого значения среднее по соответствующему столбцу и поделив результат на стандартное отклонение__

```scala
// найдем вектор средних и стандартого отклонения по каждому столбцу
val vectorMean = mean(X, Axis._0).t
val vectorStd = stddev(X, Axis._0).t * pow((convert(X.rows, Double)-1)/convert(X.rows, Double), .5)
```

```scala
// Подготовим отмаштабируемую матрица
val standartScalerMatrix = (((X.t(::, *) - vectorMean).t).t(::,*) / vectorStd).t
```

**Добавим к матрице *X* столбец из единиц. Вектор из единиц нужен для того, чтобы не обрабатывать отдельно коэффициент $w_0$ линейной регрессии.**

```scala
val onesVector = convert(DenseVector.fill(standartScalerMatrix.rows, 1), Double)
```

```scala
// Созданим новую матрицу X_train, это матрица наших признаков, для которых будут делаться предсказания
val X_train = DenseMatrix.horzcat(standartScalerMatrix, DenseMatrix.create(onesVector.length,1, onesVector.toArray))
```

**Реализуем функцию mserror - среднеквадратичную ошибку прогноза. Она принимает два аргумента - объекты y (значения целевого признака) и y_pred (предсказанные значения).**

```scala
def mseerror = (y: DenseVector[Double], y_pred: DenseVector[Double]) => ({
    sum((y - y_pred) * (y - y_pred) / convert(y.length, Double))
})
```

**Реализуем функцию *normal_equation*, которая по заданным матрицам *X* и *y* вычисляет вектор весов $w$ согласно нормальному уравнению линейной регрессии.**

```scala
def normal_equation = (X: DenseMatrix[Double], y: DenseVector[Double]) => ({
    pinv(X) * y
})
```

```scala
// Найдем наши веса
val norm_eq_weights = normal_equation(X_train(0 to 10000, ::), y(0 to 10000))
```

**Напишем функцию *linear_prediction*, которая принимает на вход матрицу *X_train* и вектор весов линейной модели *w*, а возвращает вектор прогнозов в виде линейной комбинации столбцов матрицы *X_train* с весами *w*.**

```scala
def linear_prediction = (X: DenseMatrix[Double], w: DenseVector[Double]) => ({
    X * w
})
```

```scala
// Выведем наши предсказания для каждого объекта из X_train
linear_prediction(X_train(0 to 10000, ::), norm_eq_weights)
```

**Проверим на сколько в среднем ошибается наша модель**

```scala
mseerror(y(0 to 10000), linear_prediction(X_train(0 to 10000, ::), norm_eq_weights))
```
